```python
import re
import pandas as pd

from konlpy.tag import Mecab
from konlpy.tag import Okt
from collections import Counter

#mecab=Mecab()
okt=Okt()
```

```python
df=pd.read_csv('네이버1.csv')
```

```python
# 형태소 분석
#stopwords = ['을', '를', '이', '가', '은', '는', 'null','에','에서','네','들'] #불용어 설정
tokenizer = Okt() #형태소 분석기 
token_list = []

for text in df['review']:
    txt = re.sub('[^가-힣a-z]', ' ', str(text)) #한글과 영어 소문자만 남기고 다른 글자 모두 제거
    token = tokenizer.morphs(txt) #형태소 분석
    #token =[t for t in token if len(t)>1]
    token_list.append(token)
            
token_list
```

#### 형태소 분석, 이중 리스트 하나 꺼내기

```python
# 형태소 분석, 이중 리스트 하나 꺼내기
result=[]
for t in token_list:
    for t2 in t :
        result.append(t2)

result
```

#### 한 글자 단어 제거

```python
remove_one_word = [x for x in result if len(x)>1]
len(remove_one_word)
```

#### 어미 원형 추출

```python
b=[]
for m in remove_one_word:
    m2=okt.pos(m,stem=True)
    b.append(m2)
b
```

#### 어미 원형 추출, 이중 리스트 꺼내기

```python
result=[]
for t in b:
    for t2 in t :
        result.append(t2)

result
```

#### 튜플 첫번째 값들만 출력하기

```python
first_value_tuple = [x[0] for x in result] 
first_value_tuple
```

#### value counts

```python
frequent2 = Counter(first_value_tuple).most_common()
frequent2
```

#### DataFrame으로 만들기

```python
result2 = pd.DataFrame(frequent2,columns=['word','count'])
result2
```

```python
result2.to_csv('어미원형추출_네이버1.csv')
```
